# CINEMATICA DIRECTA

```python
import math
from sympy import symbols, cos, sin, Matrix, pi

# Defino los ángulos en grados
Q1_degrees = 90
Q2_degrees = 51.6851
Q3_degrees = -54.194
Q4_degrees = -87.49

# Convierto los ángulos a radianes
Q1 = math.radians(Q1_degrees)
Q2 = math.radians(Q2_degrees)
Q3 = math.radians(Q3_degrees)
Q4 = math.radians(Q4_degrees)

# Longitudes
l1 = 96.000  # Longitud l1 en milímetros
l2 = 13.90   # Longitud l2 en milímetros
l3 = 150     # Longitud l3 en milímetros
l4 = 143.24  # Longitud l4 en milímetros
l5 = 127.42  # Longitud l5 en milímetros

# Defino las matrices de transformación usando valores numéricos
A0_1 = Matrix([
    [cos(Q1), 0, sin(Q1), l2 * cos(Q1)],
    [sin(Q1), 0, -cos(Q1), l2 * sin(Q1)],
    [0, 1, 0, l1],
    [0, 0, 0, 1]
])

A1_2 = Matrix([
    [cos(Q2), -sin(Q2), 0, l3 * cos(Q2)],
    [sin(Q2), cos(Q2), 0, l3 * sin(Q2)],
    [0, 0, 1, 0],
    [0, 0, 0, 1]
])

A2_3 = Matrix([
    [cos(Q3), -sin(Q3), 0, l4 * cos(Q3)],
    [sin(Q3), cos(Q3), 0, l4 * sin(Q3)],
    [0, 0, 1, 0],
    [0, 0, 0, 1]
])

A3_4 = Matrix([
    [cos(Q4), -sin(Q4), 0, l5 * cos(Q4)],
    [sin(Q4), cos(Q4), 0, l5 * sin(Q4)],
    [0, 0, 1, 0],
    [0, 0, 0, 1]
])

# Multiplico las matrices para obtener la matriz de transformación total
T0_4 = A0_1 * A1_2 * A2_3 * A3_4

# Evaluo la matriz con valores numéricos
T0_4_evaluada_numerica = T0_4.evalf()

# Muestro la matriz resultante en formato numérico
print("Matriz de Transformación Total (T0_4):")
print(T0_4_evaluada_numerica)
```

# CINEMATICA INVERSA

```python
import math
################################################################################################
###################################### METODO ANALITICO ########################################
################################################################################################

# Defino el vector de posicion y los vectores noa obtenidos de la cinematica directa
px =  1.53082445334849e-14
py =  250.002605553586
pz =  80.0020053361393

nx = 1.17557653498811e-21
ny = 1.91986217708551e-5
nz = -0.999999999815706
oz = 1.91986217708551e-5


l1 = 96.000 
l2 = 13.90  
l3 = 150 
l4 = 143.24 
l5 = 127.42   

# Calculo q1
q1 = math.atan2(py , px)
# Convierto a grados 
q1_degrees = math.degrees(q1)
print(f"q1 en radianes: {q1}")
print(f"q1 en grados: {q1_degrees}")


# Calculo q234 
c234 = oz
s234 = nz
#c234 = math.cos(q1)*nx + math.sin(q1)*ny
#s234 =  -math.cos(q1)*nx -  math.sin(q1)*ny

print(f"c234: {c234}")
print(f"s234: {s234}")

# Calculo el arcotangente para obtener q234
q234 = math.atan2(s234 , c234)
# Convierto a grados 
q234_degrees = math.degrees(q234)
print(f"q234 en radianes: {q234}")
print(f"q234 en grados: {q234_degrees}")


aa =  ((px / math.cos(q1)) - l5 * c234 - l2)
#((px / math.cos(q1)) - l5 * c234 - l2) = l3 * math.cos(q2) + l4 * math.cos(q2 + q3)


#(pz - l1  - l5 * s234) = l3 * math.sin(q2) + l4 * math.sin(q2 + q3)
cc =  (pz - l1  - l5 * s234)


print(f"Resultado de la primera expresión: {aa}")
print(f"Resultado de la segunda expresión: {cc}")

c = aa
d = cc

c3 = (((c**2)+(d**2)-(l4**2)-(l3**2))/(2*l4*l3))
print("c3: ",c3)

###########################################################################################
# para otra configuracion +-(math.sqrt(1-((c3)**2)))
# con + es q3 codo arriba por ende q2 me quedara para abajo ya que q2 depende de q3
# con - es q3 codo abajo por ende q2 me quedara para arriba ya que q2 depende de q3
s3 = -(math.sqrt(1-((c3)**2)))
print("s3: ",s3)

# Calculo q3
q3 = math.atan2(s3,c3)

# Convierto a grados 
q3_degrees = math.degrees(q3)
print(f"q3 en radianes: {q3}")
print(f"q3 en grados: {q3_degrees}")

r = l4*c3 + l3
s = l4*s3

# Calculo q2
q2 = math.atan2(((r*d)-(s*c)) , ((r*c)+(s*d)))

# Convierto a grados 
q2_degrees = math.degrees(q2)
print(f"q2 en radianes: {q2}")
print(f"q2 en grados: {q2_degrees}")

#q4 = math.atan2((math.cos(q1)*math.cos(q2+q3))*px + (math.sin(q1)*math.cos(q2+q3))*py + (math.sin(q2+q3))*pz - l1*math.sin(q2+q3) - l2*math.cos(q2+q3) - l3*math.cos(q3) - l4  ,l5) ####
asd = (math.cos(q1)*math.cos(q2+q3))*px + (math.sin(q1)*math.cos(q2+q3))*py + (math.sin(q2+q3))*pz - l1*math.sin(q2+q3) - l2*math.cos(q2+q3) - l3*math.cos(q3) - l4       
asd1 = l5
asd2 = asd / asd1

# Calculo q4
# Este valor de q4 no cubre todas las posibles soluciones, solo de 0 a pi
q4 = math.acos(asd2)

#Solucion a q4 para que cubra de 0 a 2pi
q4total = (2 * math.pi - q4)

#Para evitar que la articulacion rote por el por el camino mas largo, pues le resto 360 para que llegue por el camino mas corto
q4total_degrees = math.degrees(q4total) - 360
print(f"q4 en radianes: {q4total}")
print(f"q4 en grados: {q4total_degrees}")


print("------------------------------------------------------")
print(f"q1 en grados: {q1_degrees}")
print(f"q2 en grados: {q2_degrees}")
print(f"q3 en grados: {q3_degrees}")
print(f"q4 variante en grados: {q4total_degrees}")
```
